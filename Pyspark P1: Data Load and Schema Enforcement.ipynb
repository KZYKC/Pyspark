Now here lets say we need to change datatype of column Item_weight to string

_### DDL Schema method_

```python
schema_def = '''
 Item_Identifier string,
 Item_Weight string,
 Item_Fat_Content string,
 Item_Visibility double,
 Item_Type string,
 Item_MRP double,
 Outlet_Identifier string,
 Outlet_Establishment_Year integer,
 Outlet_Size string,
 Outlet_Location_Type string,
 Outlet_Type string,
 Item_Outlet_Sales double
'''
```

```python
%fs ls /Volumes/workspace/default/big_mart/
```

```python
df_new = spark.read.format('csv').schema(schema_def)\
                                 .option('header',True)\
                                 .load('dbfs:/Volumes/workspace/default/big_mart/BigMart Sales.csv')    
```

```python
df_new.printSchema()
```

```python
df_columns = df_new.columns
```

```python
df_new.display()
```

```python
schema_StructType = StructType([
                                StructField('Item_Identifier', StringType(), True),
                                StructField('Item_Weight', StringType(), True),
                                StructField('Item_Fat_Content', StringType(), True),
                                StructField('Item_Visibility', StringType(), True),
                                StructField('Item_Type', StringType(), True),
                                StructField('Item_MRP', StringType(), True),
                                StructField('Outlet_Identifier', StringType(), True),
                                StructField('Outlet_Establishment_Year', IntegerType(), True),
                                StructField('Outlet_Size', StringType(), True),
                                StructField('Outlet_Location_Type', StringType(), True),
                                StructField('Outlet_Type', StringType(), True),
                                StructField('Item_Outlet_Sales', StringType(), True)
]) #You have to pass a list of StructField objects to StructType
```

```python
df_upd = spark.read.format('csv').schema(schema_StructType)\
                                 .option('header',True)\
                                 .load('dbfs:/Volumes/workspace/default/big_mart/BigMart Sales.csv')
```
